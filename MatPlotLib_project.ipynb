```python
#Import Matplotlib 
import os 
import numpy as np 
import pandas as pd
from matplotlib import pyplot as plt
```

```python
#Creating an graph with MatPlotLib using random numbers from 0 to 500000 over a 12 month period
#generate random sales number over a 12 month period, where sales range from 0 to 500,000

a = np.random.randint(0,500000,12)
a
#plot with legends and Customization  
plt.figure(figsize=(15,10))
plt.plot(a, color = 'green', linestyle = '--',label="Sales FY20")
plt.title("Sales per Month Summary", fontsize=20, fontweight="bold")
plt.legend()
plt.xticks(range(0,len(a)),('Jan','Feb','Mar','Apr','May',"June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"))
plt.xlabel('Months')
plt.ylabel('Sales $')
plt.show()
```

```python
#generate random sales number over a 12 month period, 
#where item sales range from 0 to 40,000 (approx from 500,000/12 month period)
#where a = Pants, b = shirts, c = jackets
a = np.random.randint(0,40000,12)
b = np.random.randint(0,40000,12)
c = np.random.randint(0,40000,12)
d = (a+b+c)
#plot with legends and Customization  
plt.figure(figsize=(15,10))
plt.plot(a, color = 'green', linestyle = '--',label="Pant Sales FY20")
plt.plot(b, linewidth = 5, label="Shirt Sales FY20")
plt.plot(c, color = 'red', linestyle = '-.', label="Jacket Sales FY20")
plt.plot(d, color = 'black', label="Total Sales FY20", )
plt.title("Revenue per Month Summary")
plt.legend()
plt.xticks(range(0,len(a)),('Jan','Feb','Mar','Apr','May',"June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"))
plt.xlabel('Months')
plt.ylabel('Sales $')
plt.show()
```
